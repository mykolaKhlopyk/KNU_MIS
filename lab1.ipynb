```python
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

def task_9():
    input_list = list(map(int, input().split()))

    prime_numbers = [num for num in input_list if is_prime(num)]
    sorted_primes = sorted(prime_numbers, reverse=True)

    return sorted_primes

print(task_9()) 
```
